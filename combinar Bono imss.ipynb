```python
import pandas as pd
import openpyxl
```

```python
df1 = pd.read_excel('Clientes Combinados.xlsx')
df2 = pd.read_excel('Bono imss julio.xlsx')
```

```python
# Asegurar que no haya espacios extra en los nombres de columnas
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Copiar los nombres originales en nuevas columnas antes de normalizar
df1['NOMBRE_ORIGINAL'] = df1['NOMBRE']
df2['NOMBRE_COMPLETO_ORIGINAL'] = df2['NOMBRE COMPLETO']

# Función para normalizar nombres (maneja valores nulos)
def normalizar_nombre(nombre):
    if pd.isna(nombre):  # Evita errores con NaN
        return ''
    return ' '.join(sorted(nombre.split()))

# Aplicar normalización en las nuevas columnas
df1['NOMBRE_NORMALIZADO'] = df1['NOMBRE_ORIGINAL'].apply(normalizar_nombre)
df2['NOMBRE_NORMALIZADO'] = df2['NOMBRE_COMPLETO_ORIGINAL'].apply(normalizar_nombre)

# Hacer el merge en la columna normalizada
df_merged = df1.merge(df2, on='NOMBRE_NORMALIZADO', how='inner')

# Ver resultado
print(df_merged)

# Seleccionar solo algunas columnas de df_merged y guardarlas en un nuevo DataFrame
columnas_seleccionadas = ['ZONA_x','N_EMPRESARIO','NOMBRE_ORIGINAL', 'RFC','BONOCOP','IVA 16%','RETENCION ISR 1.25%','TOTAL A PAGAR BONOCOP']  # Cambia según lo que necesites
df_seleccionado = df_merged[columnas_seleccionadas]
# **Exportar a Excel**
archivo_salida = 'datos_cruzados_bono_imss.xlsx'
df_seleccionado.to_excel(archivo_salida, index=False, engine='openpyxl')
```
